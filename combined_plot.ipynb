```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the 6 dumped Excel files
forecast_files = {
    "SARIMA": "CPI_Forecast_SARIMA_Next_24_Months.xlsx",
    "SARIMA-Rolling": "CPI_Forecast_SARIMA_Rolling_Next_24_Months.xlsx",
    "Holt-Winters": "CPI_Forecast_HoltzWinters_Next_24_Months.xlsx",
    "Holt-Winters-Rolling": "CPI_Forecast_HoltzWinters_Rolling_Next_24_Months.xlsx",
    "Prophet": "CPI_Forecast_Prophet_Next_24_Months.xlsx",
    "Prophet-Rolling": "CPI_Forecast_Prophet_Rolling_Next_24_Months.xlsx"
}

# Read all forecast data into DataFrames
forecast_dfs = {}
for model, file_path in forecast_files.items():
    try:
        forecast_dfs[model] = pd.read_excel(file_path)
    except FileNotFoundError:
        print(f"Warning: {model} forecast file not found at {file_path}. Skipping this model.")

# Plot all forecasts in one plot
plt.figure(figsize=(12, 6))

# df.rename(columns={'Date': 'Time_Period'}, inplace=True)


for model, df in forecast_dfs.items():
    plt.plot(df['Date'], df['CPI_Forecast'], marker='o', linestyle='-', label=model)

plt.legend(fontsize=12)
plt.title('Comparison of CPI Forecasts for Next 24 Months', fontsize=14)
plt.ylabel('CPI', fontsize=12)
plt.xlabel('Time_Period', fontsize=12)
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
